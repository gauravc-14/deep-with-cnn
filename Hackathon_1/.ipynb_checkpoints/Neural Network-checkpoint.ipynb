# Importing Required Libraries

```python
import torch
from torch.utils.data import DataLoader
from torch import nn
import pandas as pd
import numpy as np
from torch.utils.data import DataLoader, Dataset
import matplotlib.pyplot as plt
```

# Reading Data

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train.describe()
```

# Preprocessing Data

```python
train.drop(axis = 1, labels = 'id', inplace = True)

y_train = train['target']
x_train = train.drop(axis = 1, labels = 'target', inplace = False)
x_train /= 200
x_train.head()
```

```python
alpha = [f"Class_{i}" for i in range(1,10)]
num = [i for i in range(0,9)]
y_train.replace(alpha,num,inplace=True)
y_train.value_counts()
```

```python
#Converting DataFrame to Tensor

train_x = torch.tensor(x_train.values.astype(np.float32)) 
train_y = torch.tensor(y_train.values.astype(np.int64))
train_x[0], train_y[0]
```

# Importing the Data into DataLoader

```python
#Combining labels and feature tensor to prepare for DataLoader Function 

class mydataset(Dataset):
    def __init__(self, data, labels):
        self.labels = labels
        self.data = data
    def __len__(self):
        return len(self.labels)
    def __getitem__(self,i):
        item = self.data[i]
        label = self.labels[i]
        return (item,label)
    
training_data = mydataset(train_x,train_y)
```

# Model

```python
class NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.sequence = nn.Sequential(
            nn.Linear(75,32),
            nn.BatchNorm1d(num_features = 32), # normalize inputs across all the batches
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.BatchNorm1d(num_features = 32),
            nn.ReLU(),
            nn.Linear(32,9),
            nn.ReLU()
        )
    def forward(self,x):
        output = self.sequence(x)
        return output
```

# Hyper-Parameters

```python
model = NN()
loss_fn = nn.CrossEntropyLoss()
epochs = 100
lr_rate = .05

optimiser = torch.optim.Adam(model.parameters(),lr = lr_rate)
```

# Training The Model

```python
def train_loop(dataloader, model, loss_fn, optimiser):
    size = len(dataloader.dataset)
    train_loss, correct = 0.0, 0
    for X, y in dataloader:
        pred = model(X)
        loss = loss_fn(pred, y)
        train_loss += loss.item() # Adding loss across all batches
        optimiser.zero_grad()
        loss.backward()
        optimiser.step()
        correct += ((nn.functional.softmax(pred, dim = 1)).argmax(1) == y).type(torch.float).sum().item()
        #Adding Accuracy across all batches
        
    train_loss /= size #mean of loss
    correct /= size  #mean of accuracy     
    print(f"Train Loss: {train_loss:.3f}")
    print(f"Train Accuracy: {correct:.3f}")
    return train_loss
```

```python
total_train_loss = []
for i in range(epochs):
    train_dataloader = DataLoader(training_data, batch_size = 40, shuffle = True)
    print(f"[=================== Epoch: {i+1} ==================]")
    train_los = train_loop(train_dataloader, model, loss_fn, optimiser)
    total_train_loss.append(train_los)
```

# Plotting Loss

```python
plt.plot(total_train_loss)
plt.xlabel("No. of Iteration")
plt.ylabel("LOSS")
plt.show
```

# Test-Set

```python
iden = test["id"].copy()
test.drop(axis = 1, labels = 'id', inplace = True)

y = nn.functional.softmax(model(torch.tensor(test.values.astype(np.float32))),dim=1)
ans = pd.DataFrame(y.detach().numpy())

lis = [f"Class_{i}" for i in range(1,10)]
ans.columns = lis
ans = pd.concat([iden,ans],axis = 1)

for i in range(1,10):
    ans[f"Class_{i}"] = ans[f"Class_{i}"].round(decimals = 4)
ans.head()
```

```python
ans.to_csv('hackathon.csv',index = False,header = True)
```
